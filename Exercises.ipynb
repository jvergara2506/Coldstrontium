```python
#7.- Exercise 3
import numpy as np
import matplotlib.pyplot as plt
def face(r,x0=0.0,y0=0.0,expression="smiling",color='b'):
    theta=np.linspace(0,2*np.pi,100)
    x=x0+r*np.cos(theta)
    y=y0+r*np.sin(theta)
    xe1=x0-r/2+r*np.cos(theta)/10
    ye1=y0+r/4+r*np.sin(theta)/10
    xe2=x0+r/2+r*np.cos(theta)/10
    ye2=y0+r/4+r*np.sin(theta)/10
    nx=x0+r*np.cos(theta)/14
    ny=y0-r/10+r*np.sin(theta)/14
    if expression=="smile":
        theta2=np.linspace(5*np.pi/4,7*np.pi/4,100)
        x2=x0+r*np.cos(theta2)/1.5
        y2=y0+r*np.sin(theta2)/1.5
    else:
        theta2=np.linspace(np.pi/4,3*np.pi/4,100)
        x2=x0+r*np.cos(theta2)/1.5
        y2=y0-r+r*np.sin(theta2)/1.5
    plt.plot(x,y,color)
    plt.plot(xe1,ye1,color)
    plt.plot(xe2,ye2,color)
    plt.plot(nx,ny,color)
    plt.plot(x2,y2,color)
plt.figure()
plt.axes().set_aspect(1)
face(2,0,0,"smile",'b')
face(1,2.5,2.5,"frown",'g')
face(1,-2.5,-2.5,"smile",'r')
plt.show()
```

```python
#7.- Exercise 4
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Latex
t,v,u=np.loadtxt('7exercise4.txt',skiprows=3,unpack=True)
def LineFit(x, y):
    xavg = x.mean()
    slope = (y*(x-xavg)).sum()/(x*(x-xavg)).sum()
    yint = y.mean()-slope*xavg
    return slope, yint
def LineFitWt(x,y,dy):
    xavg=((x/(dy*dy)).sum())/((1/(dy*dy)).sum())
    yavg=((y/(dy*dy)).sum())/((1/(dy*dy)).sum())
    slope=(((x-xavg)*y/(dy*dy)).sum())/(((x-xavg)*x/(dy*dy)).sum())
    yint=yavg-slope*xavg
    return slope,yint
def rchi2(x,y,a,b,dy=1): return (((y-a-b*x)/dy)**2).sum()
#1.-
slope1,yint1=LineFit(t,v)
slope2,yint2=LineFitWt(t,v,u)
f1=np.poly1d([slope1,yint1])
f2=np.poly1d([slope2, yint2])
print("LineFit(x,y) y=",f1)
xr2=round(rchi2(t,v,yint1,slope1,u)/(len(t)-2),3)
display(Latex(f"$X_r^2$ not weighted: {xr2}"))
print("LineFitWt(x,y,dy) y=",f2)
xr2w=round(rchi2(t,v,yint2,slope2,u)/(len(t)-2),3)
display(Latex(f"$X_r^2$ weighted: {xr2w}"))
#2.-
x=np.linspace(0,24,100)
plt.errorbar(t,v,fmt='rd',label='data',xerr=0,yerr=u,ecolor='black',ms=5,capsize=4)
plt.plot(x,f1(x),'g',label="Linear regression")
plt.plot(x,f2(x),'b',label="Linear regression with weighting")
plt.legend(loc='upper right')
plt.xlabel("time (s)")
plt.ylabel("speed (m/s)")
plt.grid(1)
print("Points with small error bars obtain higher weights, increasing their influence in determining the slope. In this case, the low uncertainty in the third value makes the slope smaller")
plt.show()
```

```python
# 7.- Exercise 5
import numpy as np
import matplotlib.pyplot as plt
t,v,u=np.loadtxt('7exercise4.txt',skiprows=3,unpack=True)
def LineFitWt(x,y,dy):
    xavg=((x/(dy*dy)).sum())/((1/(dy*dy)).sum())
    yavg=((y/(dy*dy)).sum())/((1/(dy*dy)).sum())
    slope=(((x-xavg)*y/(dy*dy)).sum())/(((x-xavg)*x/(dy*dy)).sum())
    yint=yavg-slope*xavg
    ub=1/(((x-xavg)*x/(dy*dy)).sum())
    ua=ub*(((x*x/(dy*dy)).sum())/((1/(dy*dy)).sum()))
    return slope,yint,ub,ua 
slope,yint,ub,ua=LineFitWt(t,v,u)
print("Slope: ", round(slope,2))
print("Y interception: ",round(yint,2))
print("Uncertainty slope: ", round(np.sqrt(ub),2))
print("Uncertainty y interception: ", round(np.sqrt(ua),2))
```
