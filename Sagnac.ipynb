```python
import numpy as np
import matplotlib.pyplot as plt
l=700e-9
w=100e-3
s=0.1 #+-
c=299792458 
ps=np.linspace(0,np.pi,100)
psn=ps+np.random.normal(0,ps*s)
A=(l*c*ps)/(8*np.pi*w)
plt.plot(ps,A)
plt.ylabel(r'$A (m^2)$') 
plt.gca().set_xticks([ 0,np.pi/4, np.pi/2,3*np.pi/4,np.pi])
plt.gca().set_xticklabels(['0','π/4','π/2','3π/4','π'])
plt.xlabel('Phase shift')
plt.grid(1)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from fractions import Fraction
i=0+1j
l=700e-9
w=100e-3
s=0.1 #+-
c=299792458 
k0=2*np.pi/l
k1=-k0
x=np.linspace(0,1,100)
t=np.linspace(0,1,100)
X,T=np.meshgrid(x,t)
ph=np.linspace(0,2*np.pi,9)
fig=plt.figure(figsize=(10,10))
n=1
for p in ph:
    plt.subplot(3,3,n)
    E0=np.exp(-i*(k0*X-w*T+p))
    E1=np.exp(-i*(k1*X-w*T))
    Et=abs(E0+E1)
    plt.imshow(Et,cmap='gray')
    plt.title(f"Phase shift: {Fraction(round(p/np.pi,4))} π")
    plt.xlabel('pos')
    plt.ylabel('time')
    n+=1
plt.tight_layout()
plt.show()
```
