```python
import numpy as np
import matplotlib.pyplot as plt
j=0+1j
w0=40
l=30
k=2*np.pi/l
z=np.linspace(-400, 400, 2000)
r=np.linspace(-200, 200, 2000)
Z,R=np.meshgrid(z, r)
zr=np.pi*w0*w0/l
wz = w0*np.sqrt(1+((Z)/(zr))**2)
Rz=Z*(1+((np.pi*w0*w0)/(l*Z))**2)
phiz=np.arctan(Z/zr)
E=(w0/wz)*np.exp(-(R**2)/(wz**2))*np.exp(-j*(k*Z+(k*R*R)/(2*Rz)-phiz))
fig=plt.figure(figsize=(8,8))
plt.imshow(np.abs(E.real), extent=[z.min(), z.max(), r.min(), r.max()], cmap='jet')
plt.xlabel('x (mm)')
plt.ylabel('r (mm)')
plt.grid(0)
plt.title('Intensity of a Gaussian Beam (w = 40mm,  = 30mm)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import genlaguerre

size=100
w=3*size / 4
P=5e-3
I0=2*P/(np.pi*w*w)
p=0
l=0
x=np.linspace(-size, size, 1000)
y=np.linspace(-size, size, 1000)
X,Y=np.meshgrid(x,y)
r=np.sqrt(X**2+Y**2)
phi=np.arctan2(Y, X)
rho=(2*r**2)/(w**2)
I=I0*(rho**l)*((genlaguerre(p,l)(rho))**2)*((np.cos(l*phi))**2)*np.exp(-rho)
fig=plt.figure(figsize=(6,6))
plt.imshow(I, cmap='Greens')
plt.axis('off')
plt.show()
```
